```
// Program to find power of any number.
class Exponent
{
    public int input;
    public int exponent;
    public int output = 1;

    public string ExponentUsingForLoop(int input, int exponent)
    {
        for (int i = 1; i <= exponent; i++)
        {
            output = output * input;
        }
        return $"{exponent} times {input} is {output}.";
    }

    public string ExponentUsingWhileLoop(int input, int exponent)
    {
        int i = 1;
        while (i <= exponent)
        {
            output = output * input;
            i++;
        }
        return $"{exponent} times {input} is {output}.";
    }

    public string ExponentUsingDoWhileLoop(int input, int exponent)
    {
        int i = 1;
        do
        {
            output = output * input;
            i++;
        } while (i <= exponent);
        return $"{exponent} times {input} is {output}.";
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Exponent exponent = new Exponent();
        System.Console.Write("Enter base: ");
        exponent.input = Convert.ToInt32(Console.ReadLine());
        System.Console.Write("Enter exponent: ");
        exponent.exponent = Convert.ToInt32(Console.ReadLine());
        System.Console.WriteLine(exponent.ExponentUsingForLoop(exponent.input, exponent.exponent));
        System.Console.WriteLine(
            exponent.ExponentUsingWhileLoop(exponent.input, exponent.exponent)
        );
        System.Console.WriteLine(
            exponent.ExponentUsingDoWhileLoop(exponent.input, exponent.exponent)
        );
    }
}
```

```
// Program to find maximum between three numbers.
class Maximum
{
    public double firstNumber;
    public double secondNumber;
    public double thirdNumber;

    public string MaximumNumberUsingIfElse(double valueOne, double valueTwo, double valueThree)
    {
        if (valueOne > valueTwo)
        {
            if (valueOne > valueThree)
            {
                return $"{valueOne} is greatest.";
            }
            else
            {
                return $"{valueThree} is greatest.";
            }
        }
        else
        {
            if (valueTwo > valueThree)
            {
                return $"{valueTwo} is greatest.";
            }
            else
            {
                return $"{valueThree} is greatest.";
            }
        }
    }

    public string MaximumNumberUsingTernaryOperator(
        double valueOne,
        double valueTwo,
        double valueThree
    )
    {
        return (valueOne > valueTwo)
            ? (valueOne > valueThree)
                ? $"{valueOne} is greatest."
                : $"{valueThree} is greatest."
            : (valueTwo > valueThree)
                ? $"{valueTwo} is greatest."
                : $"{valueThree} is greatest.";
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Maximum maximum = new Maximum();
        System.Console.Write("Enter first integer: ");
        maximum.firstNumber = Convert.ToDouble(Console.ReadLine());
        System.Console.Write("Enter second integer: ");
        maximum.secondNumber = Convert.ToDouble(Console.ReadLine());
        System.Console.Write("Enter third integer: ");
        maximum.thirdNumber = Convert.ToDouble(Console.ReadLine());
        System.Console.WriteLine(
            maximum.MaximumNumberUsingIfElse(
                maximum.firstNumber,
                maximum.secondNumber,
                maximum.thirdNumber
            )
        );
        System.Console.WriteLine(
            maximum.MaximumNumberUsingTernaryOperator(
                maximum.firstNumber,
                maximum.secondNumber,
                maximum.thirdNumber
            )
        );
    }
}
```

```
// Program to find all roots of a quadratic equation.
class QuadraticEquation
{
    public double a;
    public double b;
    public double c;
    public double rootOne;
    public double rootTwo;
    public double denominator;
    public double discriminant;

    public string UsingIfElse(double a, double b, double c)
    {
        denominator = 2 * a;
        discriminant = (b * b) - (4 * a * c);
        if (a == 0)
        {
            rootOne = -c / b;
            return $"The roots are Linear {Math.Round(rootOne, 2)}";
        }
        else
        {
            if (discriminant > 0)
            {
                rootOne = (-b / denominator) + (Math.Sqrt(discriminant) / denominator);
                rootTwo = (-b / denominator) - (Math.Sqrt(discriminant) / denominator);
                return $"The roots are real and distinct, {Math.Round(rootOne, 2)} and {Math.Round(rootTwo, 2)}";
            }
            else if (discriminant == 0)
            {
                rootOne = -b / denominator;
                return $"The roots are repeated, {Math.Round(rootOne, 2)}";
            }
            else
            {
                rootOne = -b / denominator;
                rootTwo = ((Math.Sqrt((4 * a * c) - (b * b))) / denominator);
                return $"The roots are imaginary and distinct, {Math.Round(rootOne, 2)}+{Math.Round(rootTwo, 2)}i and {Math.Round(rootOne, 2)}-{Math.Round(rootTwo, 2)}i";
            }
        }
    }
}

public class Program
{
    static void Main(string[] args)
    {
        QuadraticEquation quadratic = new QuadraticEquation();
        System.Console.Write("Enter a: ");
        quadratic.a = Convert.ToDouble(Console.ReadLine());
        System.Console.Write("Enter b: ");
        quadratic.b = Convert.ToDouble(Console.ReadLine());
        System.Console.Write("Enter c: ");
        quadratic.c = Convert.ToDouble(Console.ReadLine());
        System.Console.WriteLine(quadratic.UsingIfElse(quadratic.a, quadratic.b, quadratic.c));
    }
}
```

```
// Program to add and swap first and last digits of a number.
class FirstAndLast
{
    public int sum;
    public int digit;
    public int integer;
    public int temporary;
    public int lastNumber;
    public int firstNumber;

    public string FirstAndLastDigit(int input)
    {
        lastNumber = input % 10;
        digit = (int)Math.Log10(input);
        firstNumber = (int)(input / Math.Pow(10, digit));
        sum = firstNumber + lastNumber;
        temporary = firstNumber;
        firstNumber = lastNumber;
        lastNumber = temporary;
        return $"For {input} first digit is {firstNumber} and last digit is {lastNumber}\nSum of {firstNumber} and {lastNumber} is {sum}\nAfter Swapping, first digit is {firstNumber} and last digit is {lastNumber}";
    }
}

public class Program
{
    static void Main(string[] args)
    {
        FirstAndLast firstAndLast = new FirstAndLast();
        System.Console.Write("Enter an integer: ");
        firstAndLast.integer = Convert.ToInt32(Console.ReadLine());
        System.Console.WriteLine(firstAndLast.FirstAndLastDigit(firstAndLast.integer));
    }
}
```

```
// Program to calculate sum of digits of a number.
// Program to calculate product of digits of a number.
// Program to check whether a number is palindrome or not.
class Digits
{
    public int sum;
    public int input;
    public int product;
    public int reversed;
    public int remainder;
    public int temporary;

    public string Add(int input)
    {
        temporary = input;
        while (temporary != 0)
        {
            remainder = temporary % 10;
            sum = sum + remainder;
            temporary = temporary / 10;
        }
        return $"Sum of the digits of {input} is {sum}";
    }

    public string Product(int input)
    {
        input = 1;
        product = 1;
        temporary = input;
        while (temporary != 0)
        {
            remainder = temporary % 10;
            product = product * remainder;
            temporary = temporary / 10;
        }
        return $"Product of the digits of {input} is {product}";
    }

    public string Palindrome(int input)
    {
        temporary = input;
        while (temporary != 0)
        {
            remainder = temporary % 10;
            reversed = reversed * 10 + remainder;
            temporary = temporary / 10;
        }
        return (input == reversed)
            ? $"{input} is a palindrome number."
            : $"{input} is a palindrome number.";
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Digits digits = new Digits();
        System.Console.Write("Enter an integer: ");
        digits.input = Convert.ToInt32(Console.ReadLine());
        System.Console.WriteLine(digits.Add(digits.input));
        System.Console.WriteLine(digits.Product(digits.input));
        System.Console.WriteLine(digits.Palindrome(digits.input));
    }
}
```

```
// Program to find frequency of each digit in a given integer.

class Frequency
{
    public int input;

    public void FrequencyCounter(int input)
    {
        for (int i = 0; i < 10; i++)
        {
            int counter = 0;
            for (int j = input; j > 0; j = j / 10)
            {
                int remainder = j % 10;
                if (remainder == i)
                {
                    counter++;
                }
            }
            if (counter == 0)
            {
                continue;
            }
            else
            {
                System.Console.WriteLine($"The frequency of {i} is {counter}");
            }
        }
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Frequency frequency = new Frequency();
        System.Console.Write("Enter an integer: ");
        frequency.input = Convert.ToInt32(Console.ReadLine());
        frequency.FrequencyCounter(frequency.input);
    }
}
```

```
// Program to check whether a number is Armstrong number or not.
// Program to print all Armstrong numbers between 1 to n.
class Armstrong
{
    public int digit;
    public int result;
    public int endingValue;
    public int startingValue;
    public int originalNumber;

    public void ArmstrongNumber(int startingValue, int endingValue)
    {
        if (startingValue > endingValue)
        {
            int temporary = startingValue;
            startingValue = endingValue;
            endingValue = temporary;
        }
        for (int i = startingValue; i <= endingValue; i++)
        {
            digit = 0;
            result = 0;
            originalNumber = i;

            while (originalNumber != 0)
            {
                originalNumber /= 10;
                digit++;
            }

            originalNumber = i;

            while (originalNumber != 0)
            {
                int remainder = originalNumber % 10;
                result += (int)Math.Pow(remainder, digit);
                originalNumber /= 10;
            }

            if (result == i)
            {
                System.Console.Write($"{i} ");
            }
        }
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Armstrong armstrong = new Armstrong();
        System.Console.Write("Enter starting number: ");
        armstrong.startingValue = Convert.ToInt32(Console.ReadLine());
        System.Console.Write("Enter ending number: ");
        armstrong.endingValue = Convert.ToInt32(Console.ReadLine());
        System.Console.Write(
            $"Armstrong numbers between {armstrong.startingValue} and {armstrong.endingValue} are "
        );
        armstrong.ArmstrongNumber(armstrong.startingValue, armstrong.endingValue);
    }
}
```

```
// Program to check whether a number is Perfect number or not.
// Program to print all Perfect numbers between 1 to n.

class Perfect
{
    public int sum;
    public int end;
    public int start;

    public void PerfectNumbers(int start, int end)
    {
        for (int i = start; i <= end; i++)
        {
            sum = 0;

            for (int j = 1; j < i; j++)
            {
                if (i % j == 0)
                {
                    sum += j;
                }
            }

            if (sum == i)
            {
                System.Console.Write($"{i} ");
            }
        }
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Perfect perfect = new Perfect();
        System.Console.Write("Enter staring number: ");
        perfect.start = Convert.ToInt32(Console.ReadLine());
        System.Console.Write("Enter ending number: ");
        perfect.end = Convert.ToInt32(Console.ReadLine());
        System.Console.Write($"Perfect numbers between {perfect.start} and {perfect.end} are ");
        perfect.PerfectNumbers(perfect.start, perfect.end);
    }
}
```

```
// Program to print Pascal triangle up to n rows.

class Pascal
{
    public int rows;
    public int coefficient;

    public void Triangle(int rows)
    {
        for (int i = 0; i < rows; i++)
        {
            for (int j = 1; j <= rows - i; j++)
            {
                System.Console.Write(" ");
            }
            for (int k = 0; k <= i; k++)
            {
                if (k == 0 || i == 0)
                {
                    coefficient = 1;
                }
                else
                {
                    coefficient = coefficient * (i - k + 1) / k;
                }
                System.Console.Write($"{coefficient} ");
            }
            System.Console.WriteLine("\n");
        }
    }
}

public class Program
{
    static void Main(string[] args)
    {
        Pascal pascal = new Pascal();
        System.Console.Write("Enter number of rows: ");
        pascal.rows = Convert.ToInt32(Console.ReadLine());
        System.Console.WriteLine($"Pascal's triangle of {pascal.rows} rows is:");
        pascal.Triangle(pascal.rows);
    }
}
```

```
// Program to find reverse of an array.

int[] array = new int[] { 9, 8, 7, 2, 4, 3 };

for (int i = 0; i < array.Length / 2; i++)
{
    int temporary = array[i];
    array[i] = array[array.Length - i - 1];
    array[array.Length - i - 1] = temporary;
}

for (int i = 0; i < array.Length; i++)
{
    System.Console.Write($"{array[i]} ");
}
```

```
// Program to put even and odd elements of array in two separate array.

int oddIndex = 0,
    evenIndex = 0;
int[] array = new int[] { 6, 10, 5, 4, 9, 120, 4, 6, 10 };

for (int i = 0; i < array.Length; i++)
{
    if (array[i] % 2 != 0)
    {
        oddIndex++;
    }
    else
    {
        evenIndex++;
    }
}

int[] oddArray = new int[oddIndex];
int[] evenArray = new int[evenIndex];

for (int i = 0; i < array.Length; i++)
{
    if (array[i] % 2 != 0)
    {
        oddArray[--oddIndex] = array[i];
    }
    else
    {
        evenArray[--evenIndex] = array[i];
    }
}

Console.WriteLine("Odd Numbers:");
for (int i = 0; i < oddArray.Length; i++)
{
    Console.Write(oddArray[i] + " ");
}

Console.WriteLine("\n\nEven Numbers:");
for (int i = 0; i < evenArray.Length; i++)
{
    Console.Write(evenArray[i] + " ");
}
```

```
// Program to sort array elements in ascending or descending order.

class MergeSort
{
    static public void merge(int[] array, int p, int q, int r)
    {
        int i,
            j,
            k;
        int n1 = q - p + 1;
        int n2 = r - q;
        int[] L = new int[n1];
        int[] R = new int[n2];
        for (i = 0; i < n1; i++)
        {
            L[i] = array[p + i];
        }
        for (j = 0; j < n2; j++)
        {
            R[j] = array[q + 1 + j];
        }
        i = 0;
        j = 0;
        k = p;
        while (i < n1 && j < n2)
        {
            if (L[i] <= R[j])
            {
                array[k] = L[i];
                i++;
            }
            else
            {
                array[k] = R[j];
                j++;
            }
            k++;
        }
        while (i < n1)
        {
            array[k] = L[i];
            i++;
            k++;
        }
        while (j < n2)
        {
            array[k] = R[j];
            j++;
            k++;
        }
    }

    static public void mergeSort(int[] arr, int p, int r)
    {
        if (p < r)
        {
            int q = (p + r) / 2;
            mergeSort(arr, p, q);
            mergeSort(arr, q + 1, r);
            merge(arr, p, q, r);
        }
    }

    static void Main(string[] args)
    {
        int[] arr = { 76, 89, 23, 1, 55, 78, 99, 12, 65, 100 };
        int n = 10;
        Console.WriteLine("Merge Sort");
        Console.Write("Initial array is: ");
        for (int i = 0; i < n; i++)
        {
            Console.Write(arr[i] + " ");
        }
        mergeSort(arr, 0, n - 1);
        Console.Write("\nSorted Array is: ");
        for (int i = 0; i < n; i++)
        {
            Console.Write(arr[i] + " ");
        }
    }
}
```

```
// Program to sort even and odd elements of array separately.

int[] array = new int[] { 2, 3, 4, 6, 8, 5, 1, 10, 7, 9 };

for (int i = 0; i < array.Length; i++)
{
    if (array[i] % 2 == 0)
    {
        continue;
    }
    for (int j = i + 1; j < array.Length; j++)
    {
        if (array[j] % 2 == 0)
        {
            int temporary = array[i];
            array[i] = array[j];
            array[j] = temporary;
        }
    }
}

System.Console.WriteLine($"Sorted Array is\n");
for (int i = 0; i < array.Length; i++)
{
    System.Console.WriteLine($"{array[i]}");
}
```

```
// Program to left rotate an array.

int[] array = new int[] { 2, 3, 4, 6, 8, 5, 1, 10, 7, 9 };
int rotationTime = 1,
    temporary;

for (int i = 0; i < rotationTime; i++)
{
    temporary = array[0];
    for (int j = 0; j < array.Length - 1; j++)
    {
        array[j] = array[j + 1];
    }

    array[array.Length - 1] = temporary;
}

System.Console.WriteLine($"\nArray elements after {rotationTime} time(s) left rotation: ");

for (int i = 0; i < array.Length; i++)
{
    System.Console.Write($"{array[i]} ");
}
```

```
// Program to right rotate an array.

int[] array = new int[] { 2, 3, 4, 6, 8, 5, 1, 10, 7, 9 };
int j = 0,
    rotationTime = 1;

for (int i = 0; i < rotationTime; i++)
{
    int temporary = array[array.Length - 1];
    for (j = array.Length - 1; j > 0; j--)
    {
        array[j] = array[j - 1];
    }
    array[j] = temporary;
}

System.Console.WriteLine($"\nArray elements after right rotation: ");

for (int i = 0; i < array.Length; i++)
{
    System.Console.Write($"{array[i]} ");
}
```

```
// Program to merge two array to third array.

int[] inputArrayOne = new int[] { 76, 89, 23, 1, 55, 78, 99, 12, 65, 100 };
int[] inputArrayTwo = new int[] { 2, 3, 4, 6, 8, 5, 1, 10, 7, 9 };
int[] outputArray = new int[inputArrayOne.Length + inputArrayTwo.Length];

for (int i = 0; i < inputArrayOne.Length; i++)
{
    outputArray[i] = inputArrayOne[i];
}

for (int i = 0; i < inputArrayTwo.Length; i++)
{
    outputArray[i + inputArrayOne.Length] = inputArrayTwo[i];
}

System.Console.WriteLine($"The merged array");
for (int i = 0; i < outputArray.Length; i++)
{
    System.Console.Write($"{outputArray[i]} ");
}

System.Console.WriteLine($"\n\nFinal array after sorting");
for (int i = 0; i < outputArray.Length; i++)
{
    for (int j = i + 1; j < outputArray.Length; j++)
    {
        if (outputArray[i] > outputArray[j])
        {
            int temporary = outputArray[i];
            outputArray[i] = outputArray[j];
            outputArray[j] = temporary;
        }
    }
}

for (int i = 0; i < outputArray.Length; i++)
{
    System.Console.Write($"{outputArray[i]} ");
}
```

```
// Program to enter temperature in Celsius and convert it into Fahrenheit.

double celsius = 33;

double celsiusToFahrenheit = celsius * 9 / 5 + 32;
double celsiusToKelvin = 273.15 + celsius;

System.Console.WriteLine(
    $"\n{celsius} degree Celsius is equal to {Math.Round(celsiusToFahrenheit, 2)} degree Fahrenheit and {Math.Round(celsiusToKelvin, 2)} degree Kelvin\n"
);
```
